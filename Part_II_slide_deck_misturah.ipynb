## Investigation Overview


> The goal of this prsentation is to show thw factors that influence the trip dutaion of a bike sharing app: Ford GoBike. 


## Dataset Overview

> The dataset used is of Ford GoBike system data. It is a bike trip data for individual rides made in a bike-sharing system covering the greater San Francisco Bay area in February 2019. The Dataset consists of information regarding 183412 bike trips, including its timing, location of the start and end of each trip, duration of trips, user tpes and gender of users. I converted the data types of start_time and end_time to datetime formart. I also change the ids to float data type. After that I removed missing rows. I was also able to derive the weekday as well as duration of trips in minute.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv(r'C:\Users\USER\Downloads\project-template\ford_clean.csv')
```

## Visualization 1

> Subcriber user types takes most trips compare to customer user type

```python
#Lets see which user type takes the most trips

plt.figure(figsize=[8,5])
plt.bar(x = df.user_type.value_counts().keys(), height = df.user_type.value_counts() )
plt.title('Ford Go-Bike User type')
plt.xlabel('Type of User')
plt.ylabel('Number of Users')
plt.yticks(range(0,180000,20000))
plt.show()
```

```python
# Showing user type by it's percentage
plt.figure(figsize=(8,5))
user_type = df['user_type'].value_counts()
plt.pie(user_type, labels = user_type.index, startangle = 90,
        counterclock = False, autopct='%1.2f%%');
plt.title('Ford Go-Bike User type in percentage', fontsize=20)
plt.legend(['Customer','Subcriber'])
plt.show()
```

## Visualization 2

> However, customers have longer trips than subscribers

```python
plt.figure(figsize = [8, 5])
base_color = sb.color_palette()[1]
sb.boxplot(data = df, x = 'member_gender', y = 'duration_minutes', color = base_color)
plt.ylim([-5, 40])
plt.xlabel('Gender')
plt.ylabel('Duration in minutes') 
plt.title('Gender and Trip duration')
plt.show()
```

## Visualization 3
>Also, subscribers make use of this service more during th weekdays for job related activitis while customers use this service more for leisure. This could be why they take longer trips


```python
colorbar = sb.color_palette()[1]
weekday = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
g = sb.catplot(data=df, x='weekday', col="user_type", kind='count', sharey = False, color = colorbar, order = weekday)
g.set_axis_labels("Weekday", "Number of Trips")
g.set_titles("{col_name}")
g.fig.suptitle('Weekly usage of the bike share system per user type', y=1.03,);
```

## Visualization 4

> In a clearer visualisation, we can see clearly the previous conclusions.

```python
sb.pointplot(data=df, x='weekday', y='duration_minutes', hue='user_type', errwidth=0, order = weekday)
plt.xlabel('Weekday')
plt.ylabel('Duration in minutes')
plt.title('Bike Duration per day per User Type')
plt.yticks(range(0, 40, 5)) 
plt.legend(title='User Type')
plt.show()
```
